```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
data = pd.read_csv('/content/gameandgrade new.csv')


data['Grade'] = pd.to_numeric(data['Grade'], errors='coerce')

df = data[data['Grade'] > 85]
df.index
plt.figure(figsize=(10, 5))
plt.bar(df.index, df['Grade'], width=0.1)  # Use the index as the x-axis and 'Grade' as the y-axis
plt.xlabel('Index')  # Optional: label for x-axis
plt.ylabel('Grade')  # Optional: label for y-axis
plt.title('Bar Chart of Grades')  # Optional: title of the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram for 'Grade'
plt.figure(figsize=(10, 5))
plt.hist(df['Grade'], bins=10, color='g', edgecolor='black')  # You can adjust the number of bins
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Histogram of Grades')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Box plot for 'Grade'
plt.figure(figsize=(10, 5))
plt.boxplot(df['Grade'], vert=False)  # Horizontal boxplot
plt.xlabel('Grade')
plt.title('Box Plot of Grades')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot for 'Grade'
plt.figure(figsize=(10, 5))
plt.scatter(df.index, df['Grade'], color='r', marker='x')  # Red 'x' markers
plt.xlabel('Index')
plt.ylabel('Grade')
plt.title('Scatter Plot of Grades')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Line plot for 'Grade' over the index
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Grade'], marker='o', linestyle=':', color='r')  # Line plot with markers
plt.xlabel('Index')
plt.ylabel('Grade')
plt.title('Line Plot of Grades')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a categorical column for passing/failing or grades in bins
# Create a pie chart for categorical data (e.g., passing/failing grades)
passing = (df['Grade'] > 80).sum()
failing = (df['Grade'] <= 80).sum()

labels = ['Passing', 'Failing']
sizes = [passing, failing]
colors = ['#66b3ff', '#ff6666']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Passing vs. Failing Grades')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate average grade by sex
avg_grade_by_sex = data.groupby('Sex')['Grade'].mean()

# Bar plot to show average grade by sex
plt.figure(figsize=(8, 5))
avg_grade_by_sex.plot(kind='bar', color=['skyblue', 'orange'])
plt.xlabel('Sex')
plt.ylabel('Average Grade')
plt.title('Average Grade by Sex')
plt.xticks([0, 1], ['Female', 'Male'], rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Grade vs Playing Hours
plt.figure(figsize=(10, 5))
plt.scatter(data['Playing Hours'], data['Grade'], color='purple', alpha=0.5)
plt.xlabel('Playing Hours')
plt.ylabel('Grade')
plt.title('Grade vs Playing Hours')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each sex
sex_counts = data['Sex'].value_counts()

# Pie chart for the proportion of sex
plt.figure(figsize=(7, 7))
plt.pie(sex_counts, labels=['Female', 'Male'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Proportion of Sex')
# plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt

# Box plot of grades by school code
plt.figure(figsize=(10, 5))
plt.boxplot([data[data['School Code'] == code]['Grade'] for code in data['School Code'].unique()])
plt.xlabel('School Code')
plt.ylabel('Grade')
plt.title('Grade Distribution by School Code')
plt.xticks([1, 2], ['School 1', 'School 2'], rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Line plot to show the relationship between Grade and Percentage
plt.figure(figsize=(10, 5))
plt.plot(data['Grade'], data['percentage'], marker='o', color='r')
plt.xlabel('Grade')
plt.ylabel('Percentage')
plt.title('Grade vs. Percentage')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by Father Education and calculate average parent revenue
parent_revenue_by_father_education = data.groupby('Father Education')['Parent Revenue'].mean()
print(parent_revenue_by_father_education)
# # Bar plot to show average parent revenue by father education level
plt.figure(figsize=(10, 5))
parent_revenue_by_father_education.plot(kind='bar', color='lightcoral')
plt.xlabel('Father Education Level')
plt.ylabel('Average Parent Revenue')
plt.title('Average Parent Revenue by Father Education Level')
plt.xticks(rotation=45)
plt.show()
```
