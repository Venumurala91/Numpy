```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
data = sns.load_dataset('tips')  # Built-in dataset in Seaborn

# Scatter plot showing relationship between total bill and tip
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
data = sns.load_dataset('flights')

# Pivot the data to create a 2D table of passengers by month and year
pivot_data = data.pivot_table(values='passengers', index='month', columns='year')

# Heatmap of the passenger counts
sns.heatmap(pivot_data, annot=True, cmap='YlGnBu')
plt.title('Heatmap: Passengers per Month and Year')
plt.show()
```
