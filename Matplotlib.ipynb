```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/top_rated_movies(tmdb).csv')

# Categorize movies based on popularity
popularity_threshold = 50
data['popularity_category'] = ['High' if x >= popularity_threshold else 'Low' for x in data['popularity']]

# Count the number of movies in each category
popularity_counts = data['popularity_category'].value_counts()

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(popularity_counts, labels=popularity_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Movies by Popularity')
plt.show()
```

```python
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')  # Ensure 'release_date' is in datetime format
data['year'] = data['release_date'].dt.year

data.head()
yearly_avg_vote = data.groupby('year')['vote_average'].mean()

plt.plot(yearly_avg_vote.index, yearly_avg_vote.values, marker='o')
plt.title('Average Vote Per Year')
plt.xlabel('Year')
plt.ylabel('Average Vote')
plt.grid(True)
plt.show()
```

```python
plt.hist(data['vote_count'].dropna(), bins=20, edgecolor='black', alpha=0.7)
plt.title('Vote Count Distribution')
plt.xlabel('Vote Count')
plt.ylabel('Frequency')
plt.show()
```
