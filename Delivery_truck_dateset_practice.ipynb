```python
bookings_per_day = data.groupby('BookingID_Date').size()
# bookings_per_day
```

```python
# Plot the booking trend over time
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
bookings_per_day.plot(kind='line', title='Booking Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count of Market vs Regular bookings
market_regular_count = data['Market/Regular '].value_counts()

plt.figure(figsize=(8,6))
market_regular_count.plot(kind='bar', title='Market vs Regular Bookings')
plt.xlabel('Booking Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of trips for each vehicle
vehicle_trips = data['vehicle_no'].value_counts()

plt.figure(figsize=(20,6))
vehicle_trips.plot(kind='bar', title='Number of Trips by Vehicle')
plt.xlabel('Vehicle Number')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

# Count the number of trips for each vehicle
vehicle_trips = data['vehicle_no'].value_counts().reset_index()
vehicle_trips.columns = ['Vehicle Number', 'Number of Trips']

# Create an interactive bar plot
fig = px.bar(vehicle_trips,
             x='Vehicle Number',
             y='Number of Trips',
             title='Number of Trips by Vehicle',
             color='Vehicle Number',
             labels={'Vehicle Number': 'Vehicle Number', 'Number of Trips': 'Number of Trips'})

# Customize the appearance of the plot
fig.update_layout(
    xaxis_title="Vehicle Number",
    yaxis_title="Number of Trips",
    xaxis_tickangle=-45  # Rotate x-axis labels for better readability
)

# Show the plot
fig.show()
```

```python
customer_booking_count =data.groupby('customerNameCode')['BookingID'].count().reset_index()
customer_booking_count.columns=['customerNameCode','count']
customer_booking_count
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the customer booking count
plt.figure(figsize=(12,6))

sns.barplot(x='customerNameCode', y='count', data=customer_booking_count)

# Set labels and title
plt.title('Booking Count by Customer Code')
plt.xlabel('Customer Name Code')
plt.ylabel('Number of Bookings')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a Seaborn scatter plot with better styling and layout
plt.figure(figsize=(15,6))

# Adjust the scatter plot with better markers and reduced overlap
sns.set(style="whitegrid")
sns.scatterplot(x='customerNameCode', y='count', data=customer_booking_count, s=100, color='blue', edgecolor='black')

# Set title and labels with larger font sizes for better readability
plt.title('Booking Count by Customer Code', fontsize=16)
plt.xlabel('Customer Name Code', fontsize=14)
plt.ylabel('Number of Bookings', fontsize=14)

# Rotate x-axis labels to avoid overlap and ensure they fit properly
plt.xticks(rotation=45, ha='right')

# Automatically adjust the layout to avoid clipping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import plotly.express as px

# Count the number of shipments by material type
material_count = data['Material Shipped'].value_counts().reset_index()
material_count.columns = ['Material Type', 'Count']

# Create an interactive bar chart with Plotly
fig = px.bar(material_count.head(10),
             x='Material Type',
             y='Count',
             title='Top 10 Materials Shipped',
             labels={'Material Type': 'Material Type', 'Count': 'Number of Shipments'},
             text='Count',  # Adding text on the bars to show counts
             color='Material Type',  # Color each bar based on the material type for better distinction
             color_discrete_sequence=px.colors.qualitative.Set1)  # Color palette

# Customize the layout for better appearance
fig.update_layout(
    xaxis_title="Material Type",
    yaxis_title="Number of Shipments",
    xaxis_tickangle=45,  # Rotate x-axis labels for better readability
    title_x=0.5,  # Center the title
    plot_bgcolor='white',  # Set background to white for better contrast
    showlegend=False  # Hide the legend since it's not necessary for this plot
)

# Show the plot
fig.show()


```

```python
# Assuming 'Planned_ETA' is in datetime format, calculate trip duration
# Ensure 'Planned_ETA' and 'trip_start_date' are in datetime format
import matplotlib.pyplot as plt
import pandas as pd

data['Planned_ETA'] = pd.to_datetime(data['Planned_ETA'])
data['trip_start_date'] = pd.to_datetime(data['trip_start_date'])

# Calculate the trip duration in hours
data['trip_duration'] = (data['Planned_ETA'] - data['trip_start_date']).dt.total_seconds() / 3600  # in hours

# Plot trip duration distribution
plt.figure(figsize=(12,6))
plt.hist(data['trip_duration'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Trip Duration Distribution')
plt.xlabel('Trip Duration (hours)')
plt.ylabel('Frequency')
plt.show()


```
